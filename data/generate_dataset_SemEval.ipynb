```python
import json
import operator
import os
import re
import sys
import pandas as pd
```

```python
data_dir='./semeval2014/bert-pair/' 
```

```python
train = pd.read_csv(data_dir+"train_QA_M.csv",sep='\t', names=["id", "label", "sentence2", "sentence1"])
test  = pd.read_csv(data_dir+"test_QA_M.csv",sep='\t', names=["id", "label", "sentence2", "sentence1"])
```

```python
print(train.shape)
print(test.shape)
```

```python
available_labels = ["price", "anecdotes", "food", "ambience", "service"]

train['aspect'] = ""
test['aspect'] = ""

for index, row in train.iterrows():

    if "price" in row["sentence2"]:
        train.at[index,'aspect'] = "price" 

    if "anecdotes" in row["sentence2"]:
        train.at[index,'aspect'] = "anecdotes" 

    if "food" in row["sentence2"]:
        train.at[index,'aspect'] = "food" 

    if "ambience" in row["sentence2"]:
        train.at[index,'aspect'] = "ambience" 

    if "service" in row["sentence2"]:
        train.at[index,'aspect'] = "service" 

for index, row in test.iterrows():

    if "price" in row["sentence2"]:
        test.at[index,'aspect'] = "price" 

    if "anecdotes" in row["sentence2"]:
        test.at[index,'aspect'] = "anecdotes" 

    if "food" in row["sentence2"]:
        test.at[index,'aspect'] = "food" 

    if "ambience" in row["sentence2"]:
        test.at[index,'aspect'] = "ambience" 

    if "service" in row["sentence2"]:
        test.at[index,'aspect'] = "service" 
```

```python
train["sentence1"] = train["aspect"] + " " + train["sentence1"]
test["sentence1"] = test["aspect"] + " " + test["sentence1"]

# delete 
del train['sentence2']
del train['aspect']
del test['sentence2']
del test['aspect']

train.head(15)
```

```python
data_dir_out='./semeval2014/bert-sentclass-semeval2014/'

train.to_csv(data_dir_out+"train_QA_M.csv",sep='\t')
test.to_csv(data_dir_out+"test_QA_M.csv",sep='\t')
```
