```python
import json
import operator
import os
import re
import sys
import pandas as pd
```

```python
data_dir='./sentihood/bert-sentclass/'
```

```python
train = pd.read_csv(data_dir+"train_QA_M.tsv",sep='\t')
test  = pd.read_csv(data_dir+"test_QA_M.tsv",sep='\t')
dev = pd.read_csv(data_dir+"dev_QA_M.tsv",sep='\t')
```

```python
print(train.shape)
print(test.shape)
print(dev.shape)
```

#### PROCESS TRAIN 

```python
# Delete none if sentiement present.

df_train_out  = pd.DataFrame()
grouped_train = train.groupby('id')

for group_name, df_group in grouped_train:
    if (df_group['label'] == "None").all():
        df_train_out = df_train_out.append(df_group)
    else:
        for index, row in df_group.iterrows():
            if (row["label"] != "None"):
                 df_train_out = df_train_out.append(row)
        
print(df_train_out.shape)
```

```python
data_dir_out='./sentihood/bert-sentclass-processed/'
df_train_out.to_csv(data_dir_out+"processed_train_QA_M.tsv",sep='\t')
```

```python
# Delete 50% of selection (When 4 nones for the same id)

df_train_of_nones  = pd.DataFrame()
df_train_out  = pd.DataFrame()

grouped_train = train.groupby('id')

count_full_none = 0
count_full_none_selected = 0

for group_name, df_group in grouped_train:
    if (df_group['label'] == "None").all():
        count_full_none += 1
        if count_full_none % 2 == 0:
            df_train_of_nones = df_train_of_nones.append(df_group)
            count_full_none_selected += 1
    else:
        df_train_out = df_train_out.append(df_group)
        
df_train_out = df_train_out.append(df_train_of_nones)


print(count_full_none)
print(count_full_none_selected)
```

```python
print(df_train_out.shape)
```

```python
data_dir_out='./sentihood/bert-sentclass-delete50none/'
df_train_out.to_csv(data_dir_out+"train_QA_M.tsv",sep='\t')
```
