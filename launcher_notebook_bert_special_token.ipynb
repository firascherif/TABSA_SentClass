```python
from transformers import BertTokenizer
tz = BertTokenizer.from_pretrained("bert-base-cased")

special_tokens = {'additional_special_tokens': ['LOCATION1', 'LOCATION2', '[SEP1]', '[SEP2]']}
tz.add_special_tokens(special_tokens_dict=special_tokens)

#model.resize_token_embeddings(tokenizer_default.vocab_size)


sent = "I stayed in [SEP1] LOCATION1 general [SEP2] and loved the area"


# Encode the sentence
encoded = tz.encode_plus(
    text=sent,  # the sentence to be encoded
    add_special_tokens = True,  # Add [CLS] and [SEP]
    max_length = 32,  # maximum length of a sentence
    pad_to_max_length=True,  # Add [PAD]s
    return_attention_mask = True,  # Generate the attention mask
    return_tensors = 'pt',  # ask the function to return PyTorch tensors
)

print(tz.tokenize(sent))
print(encoded['input_ids'])
```
